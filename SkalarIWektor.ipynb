#### Funkcja o dwóch zmiennych, to z graficznego punktu widzenia wykres przestawiony w trójwymiarze. Funkcja o dwóch zmiennych zawiera w sobie dwie zmienne f(x, y) - w tym przypadku te zmienne to 'x' i 'y'.

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D
```

# Ustalenie funkcji. W tym przypadku:
### f(x,y) = - x^2 - y^2

```python
def f(x, y):
    return - x**2 - y**2
```

Ustalenie zakresu obliczeń.

```python
x = np.linspace(-4, 4, 10)
y = np.linspace(-4, 4, 10)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
Z = f(X,Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='jet', edgecolor = 'none')
```

# Ustalenie funkcji. W tym przypadku:
### f(x,y) = x^2 + 2xy^2

```python
def f(x, y):
    return x**2 + 2 * x * y**2
```

```python
x = np.linspace(-4, 4, 10)
y = np.linspace(-4, 4, 10)
```

```python
X, Y = np.meshgrid(x, y)

Z = f(X,Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='jet', edgecolor = 'none')
```
